```python
import py_entitymatching as em
import pandas as pd
```

```python
#This routine concatenates FirstNm, LastNm and DOB Year and created corresponding column
def transform_df(df):
    df_new = df.copy()
    df_new["full_name_dob_yr"] = df_new['given_name'] +" "+ df_new["surname"]+" "+ df_new["date_of_birth"].astype(str).str[:4]
    return df_new
```

```python
#Reading the CSV in dataframe regularly
df = pd.read_csv("people.csv", skipinitialspace=True)
df["full_name_dob_yr"] = df['given_name'] +" "+ df["surname"]+" "+ df["date_of_birth"].astype(str).str[:4]
df_a = df.copy()
df_b = df.copy()

```

```python
#Reading the sample labeled file.
G = em.read_csv_metadata('sample_label_data.csv', 
                         key='_id',
                         ltable=df_a, rtable=df_b, 
                         fk_ltable='ltable_rec_id', fk_rtable='rtable_rec_id')
```

```python
# Generate features automatically 
feature_table = em.get_features_for_matching(df_a, df_b, validate_inferred_attr_types=False)
```

```python
# Select the attrs. to be included in the feature vector table
attrs_from_table = ['ltable_full_name_dob_yr','ltable_street_number','ltable_address_1','ltable_address_2','ltable_suburb','ltable_postcode','ltable_state','ltable_date_of_birth',
                    'rtable_full_name_dob_yr','rtable_street_number','rtable_address_1','rtable_address_2','rtable_suburb','rtable_postcode','rtable_state','rtable_date_of_birth']
```

```python
# Convert the labeled data to feature vectors using the feature table
H = em.extract_feature_vecs(G, 
                            feature_table=feature_table, 
                            attrs_before = attrs_from_table,
                            attrs_after='gold',
                            show_progress=False)
```

```python
H.fillna(value=0,inplace=True)
```

```python
# Instantiate the RF Matcher
rf = em.RFMatcher()
```

```python
# Get the attributes to be projected while training
attrs_to_be_excluded = []
attrs_to_be_excluded.extend(['_id', 'ltable_rec_id', 'rtable_rec_id', 'gold'])
attrs_to_be_excluded.extend(attrs_from_table)
```

```python
# Train using feature vectors from the labeled data.
rf.fit(table=H, exclude_attrs=attrs_to_be_excluded, target_attr='gold')
```

```python
# Convert the cancidate set to feature vectors using the feature table
L = em.extract_feature_vecs(C, feature_table=feature_table,
                             attrs_before= attrs_from_table,
                             show_progress=False, n_jobs=-1)
```

```python
L.fillna(value=0,inplace=True)
```

```python
# Get the attributes to be excluded while predicting 
attrs_to_be_excluded = []
attrs_to_be_excluded.extend(['_id', 'ltable_rec_id', 'rtable_rec_id'])
attrs_to_be_excluded.extend(attrs_from_table)
```

```python
# Predict the matches
predictions = rf.predict(table=L, exclude_attrs=attrs_to_be_excluded,                          
              append=True, target_attr='predicted', inplace=False)
```

```python
predictions.head()
```

```python
# Get the attributes to be projected out
attrs_proj = []
attrs_proj.extend(['_id', 'ltable_rec_id', 'rtable_rec_id'])
attrs_proj.extend(attrs_from_table)
attrs_proj.append('predicted')

# Project the attributes
predictions = predictions[attrs_proj]
```

```python
predictions.head()
```

```python
#Considering matches for different rec_ids only.
predictions = predictions[predictions['ltable_rec_id'] != predictions['rtable_rec_id']]
```

```python
#Creating a dictionary to store rec_id, clusterID map.
id_cluster_id = {}

#initiating the cluster ID
cluster_id = 1
```

```python
#This routine is doing the cluster ID calculation
for index, row in predictions.iterrows():
    if(row['predicted'] == 1):
        if(row['ltable_rec_id'] in id_cluster_id):
            id_cluster_id[row['rtable_rec_id']] = id_cluster_id[row['ltable_rec_id']]
        elif(row['rtable_rec_id'] in id_cluster_id):
            id_cluster_id[row['ltable_rec_id']] = id_cluster_id[row['rtable_rec_id']]
        else:
            id_cluster_id[row['ltable_rec_id']] = cluster_id
            id_cluster_id[row['rtable_rec_id']] = cluster_id
            cluster_id += 1
```

```python
#Updating the Dataframe with the cluster ID using the dictionary.
df['ClusterID'] = df['rec_id'].map(id_cluster_id)
```
